```python
# 只有notRepairDamage特征为Object类型
train_df['notRepairedDamage'].value_counts()
```

```python
# '-'应该为缺失值，把他替换为na
def replaceNotRepairedDamage(df):
    df['notRepairedDamage'].replace('-', np.nan, inplace=True)
```

```python
replaceNotRepairedDamage(train_df)
replaceNotRepairedDamage(test_df)
```

```python
# 了解price的分布
train_df['price'].value_counts()
```

```python
# 使用seaborn和scipy尝试找出拟合价格的分布
import scipy.stats as st
import seaborn as sns

y = train_df['price']
plt.figure(1); plt.title('Johnson SU')
sns.distplot(y, kde=False, fit=st.johnsonsu)

plt.figure(2); plt.title('Normal')
sns.distplot(y, kde=False, fit=st.norm)

plt.figure(3); plt.title('Log Normal')
sns.distplot(y, kde=False, fit=st.lognorm)
```

```python
# 可以看出价格不符合正太分布，因此在进行回归前，必须进行转换
sns.distplot(train_df['price'])
print('Skewness: %f' % train_df['price'].skew())
print('Kurtosis: %f' % train_df['price'].kurt())
```

```python
# 查看numerical 和 categorical特征分析
numeric_features = ['power', 'kilometer', 'v_0', 'v_1', 'v_2', 'v_3', 'v_4', 'v_5', 'v_6', 'v_7', 'v_8', 'v_9', 'v_10', 'v_11', 'v_12', 'v_13','v_14' ]

categorical_features = ['name', 'model', 'brand', 'bodyType', 'fuelType', 'gearbox', 'notRepairedDamage', 'regionCode']

# 计算Numerical特征和price的correlation
numeric_features.append('price')
price_numeric = train_df[numeric_features]
correlation = price_numeric.corr()
# print(correlation)

print(correlation['price'].sort_values(ascending = False), '\n')
```

```python
# heatmap
f, ax = plt.subplots(figsize = (7,7))
plt.title('Correlation of Numeric Features with Price', y=1, size=16)
sns.heatmap(correlation, square=True, vmax=0.8)
```

```python
del price_numeric['price']
```

```python
# 可视化每个numerical特征的分布
f = pd.melt(train_df, value_vars=numeric_features)
g = sns.FacetGrid(f, col='variable', col_wrap=2, sharex=False, sharey=False)
g = g.map(sns.distplot, 'value')
```
